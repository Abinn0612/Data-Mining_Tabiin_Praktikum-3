```python
# Menggroup berdasarkan Popularity
total_per_category = df.groupby('Popularity')['Duration'].sum().sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt

# Membuat grafik batang
plt.figure(figsize=(10, 6))
total_per_category.plot(kind='bar', color='orange')

# Menambahkan judul dan label
plt.title('Total Duration per Popularity Category', fontsize=16)
plt.xlabel('Popularity', fontsize=12)
plt.ylabel('Total Duration', fontsize=12)

# Menambahkan grid dan rotasi label sumbu x
plt.xticks(rotation=45)
plt.grid(axis='y')

# Menampilkan Grafik
plt.tight_layout()
plt.show()
```

```python
# Menghitung jumlah lagu berdasarkan Artist1
artist_distribution = df['Artist1'].value_counts().head(5)  # Hanya menampilkan 5 artis teratas

# Membuat Pie Chart
plt.figure(figsize=(6, 6))
artist_distribution.plot(kind='pie', autopct='%1.1f%%', colors=plt.cm.Paired(range(len(artist_distribution))), startangle=90)

# Menambahkan judul
plt.title('Top 5 Artists by Number of Tracks', fontsize=16)

# Menampilkan plot
plt.ylabel('')  # Menghilangkan label sumbu y default
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
# Memilih hanya kolom numerik
numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns

# Min-Max Normalization
min_max_scaler = MinMaxScaler()
df_minmax = df.copy()
df_minmax[numeric_columns] = min_max_scaler.fit_transform(df[numeric_columns])

# Menampilkan hasil normalisasi MinMax Scaler
print("Data Asli:")
print(df[numeric_columns].head())

print("\nData setelah MinMax Scaling:")
print(df_minmax[numeric_columns].head())
```

```python
from sklearn.preprocessing import StandardScaler

# Inisialisasi objek StandardScaler
scaler = StandardScaler()

# Fit dan transformasi data menggunakan z-score Tranformation
z_score_data = scaler.fit_transform(df[numeric_columns])

# Hasil normasi Z-Score
print("Data Asli:")
print(df[numeric_columns].head())

print("\nData setelah Z-Score Scaling:")
print(z_score_data)
```

```python
# Menentukan scaling factor
scaling_factor = 1000

# Melakukan transformasi Decimal Scaling
scaled_data = df[numeric_columns] / scaling_factor

# Hasil normalisasi Decimal Scaling
print("Data Asli:")
print(df[numeric_columns].head())

print("\nData setelah Decimal Scaling:")
print(scaled_data.head())
```
